```python
import cv2

image = cv2.imread("69 Mustang Mach 1.png")
print("Image type", type(image))
# cv2.imshow("Image name", image)
# cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  

plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print("image shape", image_gray.shape)
plt.imshow(image_gray, cmap="gray")
```

```python
small_image = cv2.resize(image_rgb, (100, 100))
plt.imshow(small_image)
```

```python
# ksize
ksize = (15, 15) #the bigger this figure the more blurry
  
# Using cv2.blur() method 
image_blured = cv2.blur(image_rgb, ksize) 

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (15, 10))
ax1.imshow(image_rgb)
  
ax2.imshow(image_blured)
```

```python
# canny detects the edges of an object
image_edge = cv2.Canny(image_rgb, 125, 175) # two thresholds
plt.imshow(image_edge)
```

```python
import cv2 
  
# represents the top left corner of rectangle
start_point = (50, 120) # right, down
  
# represents the bottom right corner of rectangle
end_point = (960, 500)
  
# Red color in RGB
color = (255, 0, 0)
  
# Line thickness of 2 px
thickness = 5

#  used as bounding vector for image detection 
# Using cv2.rectangle() method
image_rectangle = cv2.rectangle(image_rgb, start_point, end_point, color, thickness)
  
plt.imshow(image_rectangle)
```
